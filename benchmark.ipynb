```python
import numpy as np
import jax
import jax.numpy as jnp
```

```python
# Read in the data
data = np.genfromtxt('data.csv', delimiter=',', skip_header=1)
np_x = np.arange(len(data)).astype(np.float32)
np_y = np.array(data[:,1]).astype(np.float32)
jnp_x = jnp.arange(len(data)).astype(jnp.float32)
jnp_y = jnp.array(data[:,1]).astype(jnp.float32)
```

```python
# Calculate the cubic regression with numpy
%timeit -n1 -r1 a, b, c, d = np.polyfit(np_x, np_y, 3)
```
